```python
# plot distribution of power
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'],bins=30,kde=True)
plt.title('Distribution of Generated Power(KW)')
plt.xlabel('Generated Power(KW)')
plt.ylabel('Frequency')
```

## EDA

```python
#Histograms for colunns
df[df.columns[:9]].hist(bins=30, figsize=(15,10))
plt.show()
```

```python
df[df.columns[9:18]].hist(bins=30,figsize=(15,10))
plt.show()
```

```python
df[df.columns[18:20]].hist(bins=30,figsize=(6,3))
plt.show()
```

```python
# Bivariate analysis 
#Scatter plot with target feature

numeric_cols = df.columns.drop('generated_power_kw', errors='ignore') #excluding 'generated_power_kw'
plt.figure(figsize=(15,30))
for i, column in enumerate(numeric_cols):
    plt.subplot(7,3,i+1)
    plt.scatter(df[column],df['generated_power_kw'])
    plt.title(f'{column} vs Generated Power(Kw)')
    plt.xlabel(column)
    plt.ylabel('Generated power (Kw)')
plt.tight_layout()
plt.show()    
```

```python
# Calculate correlation between features and target
df.corr()
```

```python
# Generate a heatmap to visualize correlation between input feature and target variable

plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),cmap='RdYlGn',annot=True,fmt='.2f')
plt.show()
```

```python
#outlier
plt.figure(figsize=(15,30))
for i, column in enumerate(df.
                           columns):
    plt.subplot(7,3,i+1)
    sns.boxplot(df[column])
plt.show()    
```
